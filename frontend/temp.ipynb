```python
def generate_privacy_policy(user_data, data_type_desc, usage_purpose_desc, sharing_partner_desc, protection_measure_desc, user_right_desc, transfer_measure_desc, legal_basis_desc):
    data_types = '\n '.join([f'- **{data}**: {data_type_desc[data]}' for data in user_data['data_types']])
    usage_purposes = '\n '.join([f'- **{purpose}**: {usage_purpose_desc[purpose]}' for purpose in user_data['usage_purposes']])
    sharing_partners = '\n '.join([f'- **{partner}**: {sharing_partner_desc[partner]}' for partner in user_data['sharing_partners']])
    protection_measures = '\n '.join([f'- **{measure}**: {protection_measure_desc[measure]}' for measure in user_data['protection_measures']])
    user_rights = '\n '.join([f'- **{right}**: {user_right_desc[right]}' for right in user_data['user_rights']])
    transfer_measures = '\n '.join([f'- **{measure}**: {transfer_measure_desc[measure]}' for measure in user_data['transfer_measures']])
    legal_basis = '\n '.join([f'- **{basis}**: {legal_basis_desc[basis]}' for basis in user_data['legal_basis']])

    template = f"""
    ## Privacy Policy

    ### Introduction
    Welcome to {user_data['company_name']}'s Privacy Policy. Your privacy is critically important to us. This policy explains the types of information we collect, how we use it, and the measures we take to protect it.

    ### Company Information
    **Company Name**: {user_data['company_name']}  
    **Address**: {user_data['address']}  
    **Contact Email**: {user_data['contact_email']}  

    ### Information We Collect
    We collect various types of information in connection with the services we provide, including:

    {data_types}

    ### How We Collect Information
    We collect information in the following ways:

    - **Directly from you**: When you provide it to us directly through forms, account creation, or communications.
    - **Automatically**: When you use our services, through cookies, log files, and similar technologies.
    - **From Third Parties**: When you link your account with third-party services or access our services through third-party platforms.

    ### How We Use Information
    We use the collected information for various purposes, including:

    {usage_purposes}

    ### Sharing Information
    We may share your information with the following parties:

    {sharing_partners}

    ### Cookies and Tracking Technologies
    We use cookies and similar tracking technologies to track the activity on our service and hold certain information. You can instruct your browser to refuse all cookies or to indicate when a cookie is being sent.

    ### Data Protection Measures
    We implement a variety of security measures to maintain the safety of your personal information:

    {protection_measures}

    ### Data Retention
    We will retain your personal data only for as long as is necessary for the purposes set out in this Privacy Policy. We will retain and use your data to the extent necessary to comply with our legal obligations, resolve disputes, and enforce our policies.

    ### Legal Basis for Processing
    We process your personal data based on the following legal grounds:

    {legal_basis}

    ### Your Rights
    You have the following rights regarding your personal data:

    {user_rights}

    ### International Data Transfer
    We may transfer your data internationally and ensure it is protected:

    {transfer_measures}

    ### Dispute Resolution
    If you have any complaints or concerns about our Privacy Policy or practices, please contact us at {user_data['contact_email']}. We will work with you to resolve any issues.

    ### Changes to This Privacy Policy
    We may update our Privacy Policy from time to time. We will notify you of any changes by posting the new Privacy Policy on this page. You are advised to review this Privacy Policy periodically for any changes.

    ### Contact Us
    If you have any questions about this Privacy Policy, please contact us at {user_data['contact_email']}.
    """

    return template
```
